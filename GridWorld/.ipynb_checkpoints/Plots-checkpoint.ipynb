```python
DF_c_20 = pd.read_csv("c_20/value_iteration_utilities.csv", header=None).transpose()
DF_c_1 = pd.read_csv("c_1/value_iteration_utilities.csv", header=None).transpose()
DF_c_01 = pd.read_csv("c_01/value_iteration_utilities.csv", header=None).transpose()

DF_k_10 = pd.read_csv("k_10/policy_iteration_utilities.csv", header=None).transpose()
DF_k_30 = pd.read_csv("k_30/policy_iteration_utilities.csv", header=None).transpose()
DF_k_50 = pd.read_csv("k_50/policy_iteration_utilities.csv", header=None).transpose()

# index : (Description, DataFrame, color_of_graph, scale)
DF_dict = {0 : ("DF_c_20 (Value Iteration) - 36 States", DF_c_20, "red", 1),
           1 : ("DF_c_1 (Value Iteration) - 36 States", DF_c_1, "blue", 1),
           2 : ("DF_c_01 (Value Iteration) - 36 States", DF_c_01, "green", 1),
           
           3 : ("DF_k_10 (Policy Iteration) - 36 States", DF_k_10, "red", 1), 
           4 : ("DF_k_30 (Policy Iteration) - 36 States", DF_k_30, "blue", 1),
           5 : ("DF_k_50 (Policy Iteration) - 36 States", DF_k_50, "green", 1),}
```

```python
for i, (k, v) in enumerate(DF_dict.items()):
    fig = go.Figure()
    for state_no in v[1].columns:
        fig.add_trace(go.Scatter(
            name="State " + str(state_no),
            x=v[1].index,
            y=v[1][state_no],
            mode="lines",
            showlegend=True))
    fig.update_layout(title={'text': v[0],
                             'y':0.9,
                             'x':0.45,
                             'xanchor': 'center',
                             'yanchor': 'top'},
                      xaxis_title="Iterations",
                      yaxis_title="Utility of State")
    fig.show()
```
